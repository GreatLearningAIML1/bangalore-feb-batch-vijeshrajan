### Build a DNN using Keras with `RELU` and `ADAM`

#### Load tensorflow

```
import tensorflow as tf
tf.reset_default_graph()
tf.set_random_seed(42)
import numpy as np
import pandas as pd
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers.core import Dense,Dropout,Activation,Flatten
from keras.layers.convolutional import Conv2D,MaxPooling2D
from keras.utils import np_utils
from sklearn import metrics
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
```

#### Collect Fashion mnist data from tf.keras.datasets 

```
(xtrain,ytrain),(xtest,ytest)=mnist.load_data()
```

#### Change train and test labels into one-hot vectors

```
ytrain = tf.keras.utils.to_categorical(ytrain, num_classes=10)
ytest = tf.keras.utils.to_categorical(ytest, num_classes=10)
```

#### Build the Graph

#### Initialize model, reshape & normalize data

```
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Reshape((784,),input_shape=(28,28,)))
model.add(tf.keras.layers.BatchNormalization())
```

#### Add two fully connected layers with 200 and 100 neurons respectively with `relu` activations. Add a dropout layer with `p=0.25`

```
#Hidden Layers
model.add(tf.keras.layers.Dense(200, activation='relu', name='Layer_1'))
model.add(tf.keras.layers.Dense(100, activation='relu', name='Layer_2'))

#Dropout Layers
model.add(tf.keras.layers.Dropout(0.5))

#Hidden layers
model.add(tf.keras.layers.Dense(60, activation='relu', name='Layer_3'))
model.add(tf.keras.layers.Dense(30, activation='relu', name='Layer_4'))

#Dropout layer
model.add(tf.keras.layers.Dropout(0.25))
```

### Add the output layer with a fully connected layer with 10 neurons with `softmax` activation. Use `categorical_crossentropy` loss and `adam` optimizer and train the network. And, report the final validation.

```
#Out Layer
model.add(tf.keras.layers.Dense(10, activation='softmax', name='Output'))
```

```
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```
model.summary()
```

```
#Graph Execution
model.fit(xtrain,ytrain,
          validation_data=(xtest, ytest),
          epochs= 30,
          batch_size=32)
```

```
score=model.evaluate(xtest,ytest)
```
