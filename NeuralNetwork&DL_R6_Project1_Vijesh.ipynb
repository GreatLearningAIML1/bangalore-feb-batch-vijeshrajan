2. Drop the columns which are unique for all users like IDs (5 points)

```
X = df.iloc[:, 3:13].values
y = df.iloc[:,13].values
```

```
X
```

3. Distinguish the feature and target set (5 points)

```
# We have created X & y varaiables on the above steps.
```

```
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder_X_1 = LabelEncoder()
X[:,1] = labelencoder_X_1.fit_transform(X[:,1])
labelencoder_X_2 = LabelEncoder()
X[:,2] = labelencoder_X_2.fit_transform(X[:,2])
```

```
onehotencoder = OneHotEncoder(categorical_features=[1])
X = onehotencoder.fit_transform(X).toarray()
X = X[:,1:]
```

4. Divide the data set into Train and test sets

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=7)
```

5. Normalize the train and test data (5 points)

```
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

6. Initialize & build the model (10 points)

```
# Building a Model
classifier = Sequential()
classifier.add(Dense(6, kernel_initializer = 'uniform', activation = 'relu'))
classifier.add(Dense(6, kernel_initializer = 'uniform', activation = 'relu'))
classifier.add(Dense(1, kernel_initializer = 'uniform', activation = 'sigmoid'))
```

```
# Compiling the Model
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```
classifier.fit(X_train, y_train, batch_size = 10, epochs = 100)
```

8. Predict the results using 0.5 as a threshold (Optional)

```
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
y_pred
```

9. Print the Accuracy score and confusion matrix (5 points)

```
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```
accuracy = (cm[0][0]+cm[1][1])/(cm[0][0]+cm[0][1]+cm[1][0]+cm[1][1])
accuracy
```

7. Optimize the model (Optional)

```
X_train1, X_test1, y_train1, y_test1 = train_test_split(X, y, test_size=0.2, random_state=7)
```

```
from keras.utils import np_utils
```

```
Y_train=np_utils.to_categorical(y_train1)
Y_test=np_utils.to_categorical(y_test1)
```

```
y_test.shape
```

```
X_train=np.expand_dims(X_train1,axis=2)
X_test=np.expand_dims(X_test1,axis=2)
```

```
X_train.shape
```

```
X_train.shape[1]
```

```
model1 = Sequential()
model1.add(Conv1D(filters=2,kernel_size=3,input_shape=(X_train.shape[1],X_train.shape[2]),padding='same',activation='relu'))
model1.add(MaxPooling1D(pool_size=2))
model1.add(Flatten())
model1.add(Dense(12,activation='relu'))
model1.add(Dense(8, activation='relu'))
model1.add(Dense(2, activation='sigmoid'))

```

```
# Compile model
model1.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```
# Fit the model
model1.fit(X_train, Y_train, validation_data=(X_test,Y_test), epochs=10, batch_size=5)
```

```
model1.predict(X_test)
```

```
y_predict=model1.predict_classes(X_test)
```

```
metrics.accuracy_score(y_test,y_predict)
```

```
metrics.confusion_matrix(y_test,y_predict)
```

```
cr=metrics.classification_report(y_test,y_predict)
print(cr)
```

7. Optimize the model (Optional) Attempt 2

```
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=7)
```

```
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```
classifier = Sequential()

classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 11))
classifier.add(Dropout(rate = 0.1))

classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))
classifier.add(Dropout(rate = 0.1))

classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
classifier.fit(X_train, y_train, batch_size = 10, epochs = 100)
```

```
classifier.predict(X_test)
```

```
y_predict=classifier.predict_classes(X_test)
```

```
metrics.accuracy_score(y_test,y_predict)
```

```
metrics.confusion_matrix(y_test,y_predict)
```

```
cr=metrics.classification_report(y_test,y_predict)
print(cr)
```
